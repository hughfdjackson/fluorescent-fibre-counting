```python
from data.generate import _pick_natural

def generate_test_examples(n = 100):
    for i in range(n):
        fibre_width = _pick_natural(1, 4)
        
        config = Config(
            image_dims=(512, 256),
            max_fibres = 50, min_fibres = 5,
            min_fibre_width = fibre_width, 
            max_fibre_width = fibre_width
        )
        components = gen_components(config)
        image = create_fibre_image(components, config)        
        image.save(f"images/image-num-{i}-count-{count(components)}.jpg")
```

```python
generate_test_examples(1000)
```
